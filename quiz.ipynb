```python
import sqlite3

# Connect to the database
conn = sqlite3.connect("quiz.db")
cursor = conn.cursor()

# Create the questions table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS questions (
        question_text TEXT,
        choice_1 TEXT,
        choice_2 TEXT,
        choice_3 TEXT,
        choice_4 TEXT,
        correct_answer TEXT
    )
''')

# Insert sample questions into the table
questions = [
    ("What is the capital of France?", "Paris", "London", "Berlin", "Madrid", "Paris"),
    ("What is the largest planet in our solar system?", "Jupiter", "Mars", "Saturn", "Neptune", "Jupiter"),
    ("Who painted the Mona Lisa?", "Leonardo da Vinci", "Pablo Picasso", "Vincent van Gogh", "Michelangelo", "Leonardo da Vinci"),
    ("What is the symbol for potassium?", "K", "P", "Pt", "Ka", "K")
]

cursor.executemany("INSERT INTO questions VALUES (?, ?, ?, ?, ?, ?)", questions)

# Commit the changes and close the connection
conn.commit()
conn.close()

print("quiz.db database created successfully!")
```

```python
import tkinter as tk
import sqlite3

# Connect to the database
conn = sqlite3.connect("quiz.db")
cursor = conn.cursor()

# Function to retrieve questions from the database
def get_questions():
    cursor.execute("SELECT * FROM questions")
    return cursor.fetchall()

# Function to check the user's answer
def check_answer(question, selected_answer):
    return question[5] == selected_answer

# Function to handle submitting the answer and moving to the next question
def submit_answer():
    global current_question, score

    user_answer = answer_var.get()
    question = questions[current_question]

    if check_answer(question, user_answer):
        score += 1

    current_question += 1
    if current_question < len(questions):
        show_question(current_question)
    else:
        show_result()

# Function to display the current question
def show_question(question_index):
    question_label.configure(text=questions[question_index][0])
    for i, choice in enumerate(questions[question_index][1:5]):
        radio_buttons[i].configure(text=choice, value=choice)
    answer_var.set('')

# Function to display the quiz result
def show_result():
    result_text = f"Quiz complete!\nYour score: {score}/{len(questions)}"
    result_label.configure(text=result_text)
    question_frame.pack_forget()
    result_frame.pack()

# Create the main window
root = tk.Tk()
root.title("Quiz Game")

# Create frames for question and result display
question_frame = tk.Frame(root)
result_frame = tk.Frame(root)

# Get questions from the database
questions = get_questions()
current_question = 0
score = 0

# Create GUI elements for the question frame
question_label = tk.Label(question_frame, text="")
question_label.pack(pady=20)

answer_var = tk.StringVar()
radio_buttons = []
for i in range(4):
    radio_button = tk.Radiobutton(question_frame, variable=answer_var)
    radio_button.pack()
    radio_buttons.append(radio_button)

submit_button = tk.Button(question_frame, text="Submit", command=submit_answer)
submit_button.pack(pady=10)

# Create GUI elements for the result frame
result_label = tk.Label(result_frame, text="")
result_label.pack(pady=20)

# Show the first question
show_question(current_question)

# Pack the frames into the main window
question_frame.pack(padx=50, pady=50)
result_frame.pack(padx=50, pady=50)

# Start the main event loop
root.mainloop()

# Close the database connection
conn.close()
```
